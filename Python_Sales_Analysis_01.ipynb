```python
# Filling missing values
data['Sales_fillNA']=data['Sales'].fillna(data['Sales'].mean())
```

```python
# Rename Columns
data.rename(columns={'Transaction Date':'Sales Date'})
```

```python
# change the data type
data['Units']=data['Units'].fillna(data['Units'].mean()).astype(int)
```

```python
data.dtypes
```

```python
data['New Price']=data['Price'].apply(lambda x:x*2)
```

```python
# modify the existing price data
print(data[['Price', 'New Price']].head())
```

```python
# aggregating and grouping
grouped_mean = data.groupby('Product')['Sales_fillNA'].mean()
print(grouped_mean)
```

```python
data.groupby(['Product','Coupon Type'])['Sales_fillNA'].sum()
```

```python
data.groupby(['Product','Coupon Type'])['Sales_fillNA'].mean()
```

```python
grouped_agg = data.groupby('Coupon Type')['Sales_fillNA'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
# merging and joining dataframes
df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'],'Value1':[4,5,6]})
df1
```

```python
df2
```

```python
# Merge Dataframes on the key columns
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
